#step 6: Missing Value Treatment

```python
missing_values_count = df.isnull().sum()
print("Number of missing values in each column:")
print(missing_values_count)
```

```python
method='mean'
if method == 'mean':
  df = df.fillna(df.mean())
elif method == 'median':
  df = df.fillna(df.median())
elif method == 'mode':
  df = df.fillna(df.mode().iloc[0])
else:
  raise ValueError("Invalid imputation method. Choose from 'mean', 'median', or 'mode'.")

print(df)
```

```python
threshold=0.95
corr_matrix = df.corr().abs()

mask = (corr_matrix >= threshold) & (corr_matrix < 1)

features_to_keep = [column for column in mask.columns if not any(mask[column])]

df = df[features_to_keep]


print(df)
```

#step 8: Multicollinearity (VIF > 5)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = df.dropna(axis=1, how='all')

if len(df) == 0 or len(df) == 1:
    print("multicollinearity is absent")
else:
    df = df.loc[:, (df != df.iloc[0]).any()]

    df = df.replace([np.inf, -np.inf], np.nan).dropna(axis=1, how='all')


    variables_to_remove = []

    for i in range(df.shape[1]):
        vif = variance_inflation_factor(df.values, i)
        if vif > 5:
            variables_to_remove.append(df.columns[i])

    df = df.drop(columns=variables_to_remove)

    if len(variables_to_remove) > 0:
        print("Removed variables with VIF greater than 5")
    else:
        print("multicollinearity is absent")

print("DataFrame after removing high VIF variables:")
print(df)
```
